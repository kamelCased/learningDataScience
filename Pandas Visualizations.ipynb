# Pandas Built-In Data Visualizations

For in-depth analyses, use seaborn. For quick data visualizations, use Pandas.

```python
# imports
import numpy as np
import pandas as pd
import seaborn as sns
sns.set_style('darkgrid')
%matplotlib inline
```

```python
# CSV data upload
df1 = pd.read_csv('files/df1',index_col=0)
df1.head()
```

### Histograms

You can create a histogram of a column with `hist`

```python
# Method 1: Create a histogram for the A column
df1['A'].hist()
```

```python
# Method 2: Create a plot with type histogram
df1['A'].plot(kind='hist')
```

```python
# Method 3 (most common for all plots): Create a histogram plot
df1['A'].plot.hist()
```

##### Adding histogram kwargs
You can specify histogram parameters such as the number of bins...

```python
# Create a histogram with 30 bins
df1['A'].plot.hist(bins=30)
```

### Scatter Plots

```python
# Create a scatte plot
df1.plot.scatter(x='D',y='B')
```

###### You can specify the color of the points to represent the value on a third column

```python
# Create a scatter plot with color representation for another column
df1.plot.scatter(x='D',y='B',c='C',cmap='coolwarm') # Data point colors represent their values on c
```

##### You can specify the size of the data points to represent their value on a third column

```python
# Create a scatter plot with variable point sizes
df1.plot.scatter(x='D',y='B',s=df1['C']*50,edgecolors='k') # Note: You have to pass in the actual df column (not just the name) and multiply to make the poitns visible
```

### Box Plots
