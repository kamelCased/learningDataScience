# MatPlotLib 

* MatPlotlib is the most popular plotting library for Python

* MatPlotLib gives you complete control over every aspect of a figure or plot

* MatPlotLib is designed to have a very similar feel to Matlab's graphical plotting

* Works very well with Pandas and NumPy arrays

## How to use MatPlotLib to create figures

```python
# Import pyplot 
import matplotlib.pyplot as plt
# only for Jupyter notebook, to visualize plots in notebook
%matplotlib inline
```

### Functional Method for plotting figures with MatPlotLib

```python
# Create 2 numpy arrays of data
import numpy as np
x = np.linspace(0,5,11) # x has 11 linearly spaced points between 0 and 5
y = x ** 2 # y is the squared value of x 
# Plot the figure (functional method)
plt.plot(x,y)
# plt.show() # If not in Jupyter notebook you would run this line to show the plot
```

#### Labelling the plot

```python
plt.plot(x,y)
# Add the x label
plt.xlabel('x')
# Add the y label
plt.ylabel('y = x ** 2')
# Add the plot title
plt.title('Quadratic Function')
```

#### Putting multi-plots on the same canvas

use `plt.subplot` to create mutliple plots on the same canvas 

```python
# plt.subplot(rows, cols, currentOneToEdit)

# Edit the first subplot 
plt.subplot(1,2,1)
# Create the plot
plt.plot(x,y,'b')

# Edit the second subplot
plt.subplot(1,2,2)
# Create the second plot
plt.plot(y,x,'r')
```

### Object Oriented Method for plotting figures with MatPlotLib

Gives a lot more control over the plot

```python
# Create a figure object 
fig = plt.figure()
# Add axes to the figure object. Parameter: [left,bottom,width,height]
axes = fig.add_axes([0.1,0.1,0.8,0.8])
# Plot the desired function on the axes
axes.plot(x,y)
# Set the x label
axes.set_xlabel('x')
# Set the y label
axes.set_ylabel('y = x ** 2')
# Set the plot title
axes.set_title('Quadratic Function')
```

#### Putting multi-plots on the same canvas

The object oriented method allows us much more control over the location of the subplots. 

```python
#### Putting multi-plots on the same canvas

# Create the figure
fig = plt.figure()

# Add first plot 
axes1 = fig.add_axes([0.5,0,0.4,0.4])
axes1.plot(x,y)
axes1.set_title('x vs y')

# Add second plot 
axes2 = fig.add_axes([0.5,0.5,0.4,0.4])
axes2.plot(y,y)
axes2.set_title('y vs y')

# Add third plot
axes3 = fig.add_axes([0,0,0.4,0.4])
axes3.plot(x,x)
axes3.set_title('x vs x')

# Add fourth plot
axes4 = fig.add_axes([0,0.5,0.4,0.4])
axes4.plot(y,x)
axes4.set_title('y vs x')
```

#### Embedded (insert) plots

```python
# Create the figure
fig = plt.figure()

# Add first plot 
axes1 = fig.add_axes([0,0,1,0.8])
axes1.plot(x,y)
axes1.set_title('x vs y')

# Add second plot (embedded in first one)
axes2 = fig.add_axes([0.1,0.2,0.3,0.3])
axes2.plot(y,x)
axes2.set_title('y vs x')
```

#### Multi-plots with subplots

Easier than manually specifying subplot locations

```python
# 1. Create a  canvas with 2rowsx2cols subplots 
fig,axes = plt.subplots(nrows=2,ncols=2)

# 2. Index the axes list to plot on the specific subplot 

# First Plot
axes[0,0].plot(x,y)
axes[0,0].set_title('x vs. y')

# Second Plot
axes[0,1].plot(x,x)
axes[0,1].set_title('x vs. x')

# Third Plot
axes[1,0].plot(y,y)
axes[1,0].set_title('y vs. y')

# Fourth Plot
axes[1,1].plot(y,x)
axes[1,1].set_title('y vs. x')

# Get rid of overlap
fig.tight_layout()
```

```python
# you can also iterate through the axes

# axis data
x = [np.linspace(0,5,11), np.linspace(0,5,11) ** 2, np.linspace(0,5,11) ** 2, np.linspace(0,5,11), np.linspace(0,5,11)]

# 1. Create a  canvas with 2rowsx2cols subplots 
fig,axes = plt.subplots(nrows=1,ncols=4)

# 2. Iterate through the axes list to plot on the specific subplot 
i = 0
for axis in axes:
    axis.plot(x[i],x[i+1])
    i += 1
```

### Figure Size and Resolution

### Figure Size and DPI with Subplots

```python
# figsize = (widthInches,heightInches), dpi = dotsPixelsPerInch
fig,axes = plt.subplots(1, 2, figsize=(3,2), dpi=100) 

# Plot the figure with specified size and resolution
i = 0
for axis in axes:
    axis.plot(x[i],y)
    i += 1
    
# Remove overlap
fig.tight_layout()
```

## Saving Figures to a File with `savefig`

```python
# Create the figure
fig,axes = plt.subplots(3,2,figsize=(5,3),dpi=150)
# Plot 1 
axes[0,0].plot(x[0],np.sin(x[0])) 
axes[0,0].set_title('x vs. sin(x)')
# Plot 2
axes[0,1].plot(x[0],x[0] ** 3) 
axes[0,1].set_title('x vs. x ** 3')
# Plot 3 
axes[1,0].plot(x[0],np.log(x[0])) 
axes[1,0].set_title('x vs. log(x)')
# Plot 4
axes[1,1].plot(x[0],x[0] ** 1/2) 
axes[1,1].set_title('x vs. sqrt(x)')
# Plot 5
axes[2,0].plot(x[0],np.exp(x[0])) 
axes[2,0].set_title('x vs. e ** x')
# Plot 6
axes[2,1].plot(x[0],np.tan(x[0])) 
axes[2,1].set_title('x vs. tan(x)')

#Cleaner look
fig.tight_layout()

# Save the figure. Syntax: fig.savefig('path/fileName.format',)
fig.savefig('files/matplotlib.png')
```

### Multiple functions on the same plot

```python
# Create the figure
fig,ax = plt.subplots()
# Plot 1st function
ax.plot(x[0],y)
# Plot 2nd function
ax.plot(x[0],np.sin(x[0]))
# Plot 3rd function ...
ax.plot(x[0],np.exp(x[0])-30) # Problem: Just from looking at the plot, we can't tell which function is represented by which curve
```

### Legends

Legends can help you provide descriptions to curves for better comprehensibility

```python
# Step 1: Label the curves

# Create the figure
fig,ax = plt.subplots()
# Plot 1st function
ax.plot(x[0],y,label='X squared')
# Plot 2nd function
ax.plot(x[0],np.sin(x[0]),label='sin(x)')
# Plot 3rd function ...
ax.plot(x[0],np.exp(x[0])-30,label='exp(x)') 

# Step 2: Add the legend
ax.legend() # Plot is now much more intelligible
```

###### Note on choosing legend location

If the legend obstructs the view you can add the parameter loc=0 to give it a better location

```python
# Bad location example

# Create the figure
fig,ax = plt.subplots()
# Plot 1st function
ax.plot(x[0],y,label='X squared')
# Plot 2nd function
ax.plot(x[0],np.sin(x[0]),label='sin(x)')
# Plot 3rd function ...
ax.plot(x[0],np.exp(x[0])-30,label='exp(x)') 

# Step 2: Add the legend
ax.legend(loc=7) 
```

```python
# Good legend location example

# Create the figure
fig,ax = plt.subplots()
# Plot 1st function
ax.plot(x[0],y,label='X squared')
# Plot 2nd function
ax.plot(x[0],np.sin(x[0]),label='sin(x)')
# Plot 3rd function ...
ax.plot(x[0],np.exp(x[0])-30,label='exp(x)') 

# Step 2: Add the legend
ax.legend(loc=0) 
```

### Plot Appearance

#### Setting colors

```python
# Create functions
x = np.linspace(0,10,250)
y = [np.cos(x),x ** 0.5, x ** 3 - x ** 2 - 12] 
# Create plots
fig,ax = plt.subplots(1,3)
# Set colors
ax[0].plot(x,y[0],color='orange') # You can set the color by name
ax[1].plot(x,y[1],color="#E42C09") # You can set the color by hex code
ax[2].plot(x,y[2],color='b') # You can set the color by first letter
```

#### Line width

```python
# Create functions
x = np.linspace(0,10,250)
y = [np.cos(x),x ** 0.5, x ** 3 - x ** 2 - 12] 
# Create plots
fig,ax = plt.subplots(1,3)
# Set line width (lw)
ax[0].plot(x,y[0],linewidth=0.5) # Smaller line width
ax[1].plot(x,y[1],lw=1) # Normal line width
ax[2].plot(x,y[2],lw=2) # Bigger line width
```

#### Line Transparency

```python
# Create functions
x = np.linspace(0,10,250)
y = [np.cos(x),x ** 0.5, x ** 3 - x ** 2 - 12] 
# Create plots
fig,ax = plt.subplots(1,3)
# Set transparency (alpha)
ax[0].plot(x,y[0],lw=2,alpha=0.2) # Low alpha (high transparency)
ax[1].plot(x,y[1],lw=2,alpha=0.5) # Medium Alpha (some transparency)
ax[2].plot(x,y[2],lw=2,alpha=1) # High alpha (no transparency)
```

#### Line Style

```python
# Create functions
x = np.linspace(0,10,250)
y = [np.cos(x),x ** 0.5, x ** 3 - x ** 2 - 12] 
# Create plots
fig,ax = plt.subplots(1,3)
# Set line style (ls)
ax[0].plot(x,y[0],linestyle='-.') # dashed dotted
ax[1].plot(x,y[1],ls='--') # dashed
ax[2].plot(x,y[2],ls=':') # dotted
```

#### Markers

```python
# Create functions
x = np.linspace(0,10,25)
y = [np.cos(x),x ** 0.5, x ** 3 - x ** 2 - 12] 
# Create plots
fig,ax = plt.subplots(1,3)
# Set markers
ax[0].plot(x,y[0],marker='o') # cirle markers
ax[1].plot(x,y[1],marker='+') # plus markers
ax[2].plot(x,y[2],marker='*') # star markers
```

#### More marker customizations

###### You can modify the marker size

```python
# Create functions
x = np.linspace(0,10,15)
y = [np.cos(x),x ** 0.5, x ** 3 - x ** 2 - 12] 
# Create plots
fig,ax = plt.subplots(3,1)
# Set marker size
ax[0].plot(x,y[0],marker='o',markersize=6) # normal markers
ax[1].plot(x,y[1],marker='o',markersize=2) # small markers
ax[2].plot(x,y[2],marker='o',markersize=10) # big markers
```

###### You can modify the marker face (fill) color

```python
# Create functions
x = np.linspace(0,10,15)
y = [np.cos(x),x ** 0.5, x ** 3 - x ** 2 - 12] 
# Create plots
fig,ax = plt.subplots(3,1)
# Set marker face color
ax[0].plot(x,y[0],marker='o',markersize=10,markerfacecolor='y') # yellow (using first letter)
ax[1].plot(x,y[1],marker='o',markersize=10,markerfacecolor='#53427A') # purple (using hex)
ax[2].plot(x,y[2],marker='o',markersize=10,markerfacecolor='orange') # orange (using name)
```

##### You can change the marker edge width

```python
# Create functions
x = np.linspace(0,10,15)
y = [np.cos(x),x ** 0.5, x ** 3 - x ** 2 - 12] 
# Create plots
fig,ax = plt.subplots(3,1)
# Set marker edge width
ax[0].plot(x,y[0],marker='o',markersize=10,markerfacecolor='y',markeredgewidth=2) # normal(normal is 1)
ax[1].plot(x,y[1],marker='o',markersize=10,markerfacecolor='#53427A',markeredgewidth=0.1) # thin 
ax[2].plot(x,y[2],marker='o',markersize=10,markerfacecolor='orange',markeredgewidth=5) # thick
```

```python
##### You can change the marker edge color

# Create functions
x = np.linspace(0,10,15)
y = [np.cos(x),x ** 0.5, x ** 3 - x ** 2 - 12] 
# Create plots
fig,ax = plt.subplots(3,1)
# Set marker edge color
ax[0].plot(x,y[0],marker='o',markersize=10,markerfacecolor='y',markeredgewidth=4,markeredgecolor='purple') # purple
ax[1].plot(x,y[1],marker='o',markersize=10,markerfacecolor='#53427A',markeredgewidth=4,markeredgecolor='#AAAAAA') # grey
ax[2].plot(x,y[2],marker='o',markersize=10,markerfacecolor='orange',markeredgewidth=4,markeredgecolor='k') # black
```

#### Bounded axis ranges

```python
##### You can change the marker edge color

# Create functions
x = np.linspace(0,10,50)
y = [np.cos(x),x ** 0.5, x ** 3 - x - 12] 
# Create plots
fig,ax = plt.subplots(3,1)
ax[0].plot(x,y[0])
ax[1].plot(x,y[1])
ax[2].plot(x,y[2])
# Set the axis bounds
for a in ax: # note: here I am bounding all axes to the same limits (you don't have to)
    a.set_xlim([0,10]) # x axis limits
    a.set_ylim([-5,5]) # y axis limits
```
