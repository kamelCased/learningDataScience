## Data Prep

### Split to Train and Test 

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('gesture', axis=1).values
y = df['gesture'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

### Scale Features

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

### One hot encode target classes

```python
from tensorflow.keras.utils import to_categorical
```

```python
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

## Modelling 

### Create Callbacks

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping, TensorBoard
from datetime import datetime
```

```python
early_stop = EarlyStopping(monitor='val_loss', 
                           patience=25,
                           verbose=1,
                           mode='min')
```

```python
time = datetime.now().strftime("%Y-%m-%d--%H%M")
board = TensorBoard(
    log_dir = f'files/logs/fit/muscle/{time}', # where to log
    histogram_freq = 1, # log a weight hist every epoch
    write_graph = True, # log graphs
    write_images = True, # log images
    update_freq = 'epoch', # update every epoch
    profile_batch = 2, #  use 2nd batch as profile
    embeddings_freq = 1 # log embeddings every epoch
)
```

### Build model

```python
X_train.shape
```

```python
model = Sequential()

model.add(Dense(64, activation='relu'))
model.add(Dropout(0.3))

model.add(Dense(32, activation='relu'))
model.add(Dropout(0.4))

model.add(Dense(16, activation='relu'))

model.add(Dense(4, activation='softmax'))
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

### Train model

```python
model.fit(X_train, y_train,
          batch_size=256,
          epochs=600,
          callbacks=[early_stop, board],
          validation_data=(X_test, y_test)
     )
```

### Monitor training history

#### In notebook

```python
training_stats = pd.DataFrame(model.history.history)
training_stats.plot()
```

### Make predictions

```python
y_test = np.argmax(y_test, axis=1)
pred = model.predict_classes(X_test)
```

### Evaluate model performance

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
sns.heatmap(confusion_matrix(y_test, pred), fmt='g', annot=True)
```

```python
# 77% accuracy, 77% precision, 77% recall, 76% f1 score
print(classification_report(y_test, pred))
```
