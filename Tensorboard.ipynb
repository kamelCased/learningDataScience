# Tensorboard

Tensorboard is a visualization tool from Google designed to work in conjunction with TensorFlow to visualize various aspects of your model.

```python
import numpy as np
import pandas as pd
```

## Get data

```python
df = pd.read_csv('files/cancer_classification.csv')
df.head()
```

## Train test split

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('benign_0__mal_1', axis=1).values
y = df['benign_0__mal_1'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

## Scaling Data

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

## Modelling 

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping, TensorBoard # import tensorboard
```

```python
early_stop = EarlyStopping(patience=20, verbose=1)
```

### Create the TensorBoard callbacks

This callback can log events for TensorBoard, including:
* Metrics summary plots
* Training graph visualization
* Activation histograms 
* Sampled profiling

```python
pwd
```

```python
from datetime import datetime
time = datetime.now().strftime("%Y-%m-%d--%H%M")
```

```python
board = TensorBoard(
    log_dir = f'files/logs/fit/{time}', # where to log
    histogram_freq = 1, # log a weight hist every epoch
    write_graph = True, # log graphs
    write_images = True, # log images
    update_freq = 'epoch', # update every epoch
    profile_batch = 2, #  use 2nd batch as profile
    embeddings_freq = 1 # log embeddings every epoch
)
```

### Build the model

```python
model = Sequential()

model.add(Dense(30, activation='relu'))
model.add(Dropout(0.2))

model.add(Dense(15, activation='relu'))
model.add(Dropout(0.4))

model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy')
```

### Train the model

```python
model.fit(X_train, y_train,
          epochs = 600, 
          validation_data = (X_test, y_test), 
          verbose = 1, 
          callbacks = [early_stop, board] # Add tensorboard to callbacks
         )
```
