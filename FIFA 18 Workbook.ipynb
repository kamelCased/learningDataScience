### Best Shot Stoppers Overall
1. Neuer
2. De Gea
3. Handanovic

```python
# weights for calculating shot stopping abilities
a = 0.1
b = 0.25
c = 0.5
d = 1
```

```python
# Create shot stopper score for every goal keeper
fifa['GK shotstopper'] = (c*fifa['Reactions'] + b*fifa['Composure'] + a*fifa['Sprint speed'] + a*fifa['Strength'] + c*fifa['Jumping'] + c*fifa['GK positioning'] + d*fifa['GK diving'] + d*fifa['GK reflexes'] + b*fifa['GK handling']) / (2*a + 2*b + 3*c + 2*d)
# Get the top 3 shot stoppers
fifa['Name'].loc[fifa['GK shotstopper'].dropna().sort_values(ascending=False).head(3).index]
```

### Best Sweeper Keepers Overall
1. Neuer
2. De Gea
3. Ter Stegen

```python
# Create sweeper score for every goal keeper
fifa['GK sweeper'] = (c*fifa['Reactions'] + b*fifa['Composure'] + b*fifa['Sprint speed'] + a*fifa['Short passing'] + b*fifa['Long passing'] + b*fifa['Jumping'] + d*fifa['GK positioning'] + b*fifa['GK diving'] + d*fifa['GK reflexes'] + c*fifa['GK handling'] + d*fifa['GK kicking'] + c*fifa['Vision']) /  (1*a + 5*b + 3*c + 3*d)
# Get the top 3 sweepers
fifa['Name'].loc[fifa['GK sweeper'].dropna().sort_values(ascending=False).head(3).index]
```

### Best Centre Backs

1. Thiago Silva
2. Sergio Ramos
3. Giorgio Chiellini

```python
# Create defense score for all players
fifa['Centre Back'] = (d*fifa['Interceptions'] + b*fifa['Jumping'] + d*fifa['Marking'] + c*fifa['Positioning'] + c*fifa['Reactions'] + d*fifa['Sliding tackle'] + d*fifa['Strength'] + d*fifa['Standing tackle'] + c*fifa['Agility'] + c*fifa['Aggression'] + a*fifa['Acceleration'] + b*fifa['Vision'] + a*fifa['CB']) /  (2*a + 2*b + 4*c + 5*d)
# Get the top 3 centre backs
fifa.loc[fifa['Centre Back'].dropna().sort_values(ascending=False).index][(fifa['Preferred Positions'] == 'CB') | (fifa['Preferred Positions'] == 'RCB') | (fifa['Preferred Positions'] == 'LCB')]['Name'].head(3)
```

### Best Wing Backs
1. Filipe Luis
2. Jordi Alba
3. Marcelo

```python
# Create wing back score for all players
fifa['Wing Back'] = (c*fifa['Sprint speed'] + b*fifa['Aggression'] + a*fifa['Agility'] + d*fifa['Crossing'] + b*fifa['Dribbling'] + b*fifa['Balance'] + d*fifa['Interceptions'] + d*fifa['Long passing'] + d*fifa['Marking'] + c*fifa['Positioning'] + d*fifa['Sliding tackle'] + c*fifa['Vision'] + d*fifa['Stamina'] + d*fifa['Standing tackle'] + d*fifa['Stamina']) /  (1*a + 3*b + 3*c + 8*d)
# Get the top 3 wing backs
fifa.loc[fifa['Wing Back'].dropna().sort_values(ascending=False).index][(fifa['Preferred Positions'] == 'RWB') | (fifa['Preferred Positions'] == 'RB') | (fifa['Preferred Positions'] == 'LWB') | (fifa['Preferred Positions'] == 'LB')]['Name'].head(3)
```

### Best Defending Midfielders
1. Radja Nainggolan
2. N'golo Kante
3. Arturo Vidal

```python
# Create score
fifa['Beast'] = (d*fifa['Aggression'] + d*fifa['Balance'] + c*fifa['Composure'] + d*fifa['Interceptions'] + c*fifa['Marking'] + d*fifa['Positioning'] + b*fifa['Short passing'] + b *fifa['Shot power'] + c*fifa['Sliding tackle'] + c*fifa['Stamina'] + d*fifa['Standing tackle'] + d*fifa['Strength']) / (2*b + 4*c + 6*d)
# Get the top 3 defensive midfielders
fifa['Name'].loc[fifa['Beast'].dropna().sort_values(ascending=False).head(3).index]
```

### Best Attacking Midfielders
1. Leo Messi
2. Kevin De Bruyne
3. Toni Kroos

```python
# Create score
fifa['Playmaker'] = (d*fifa['Ball control'] + d*fifa.Dribbling + a*fifa.Marking + d*fifa.Reactions + d*fifa.Vision + c*fifa.Positioning + d*fifa.Crossing + d*fifa['Short passing'] + d*fifa['Long passing'] + c*fifa.Curve + b*fifa['Long shots'] + c*fifa['Free kick accuracy']) / (1*a + 1*b + 2*c + 8*d)
# Get the top 3 attacking midfielders
fifa['Name'].loc[fifa['Playmaker'].dropna().sort_values(ascending=False).head(3).index]
```

### Best Wingers 
1. Cristiano Ronaldo
2. Leo Messi
3. Gareth Bale

```python
# Create score
fifa['Winger'] = (d*fifa['Acceleration'] + c*fifa['Agility'] + c*fifa['Ball control'] + c*fifa['Crossing'] + d*fifa['Curve'] + d*fifa['Dribbling'] + c*fifa['Finishing'] + d*fifa['Long shots'] + a*fifa['Positioning'] + a*fifa['Reactions'] + b*fifa['Short passing'] + d*fifa['Shot power'] + d*fifa['Sprint speed'] + c*fifa['Stamina'] + c*fifa['Vision'] + b*fifa['Volleys']) / (2*a + 2*b + 6*c + 6*d)
# Get the top 3 wingers
fifa['Name'].loc[fifa['Winger'].dropna().sort_values(ascending=False).head(3).index]
```

### Best Strikers
1. Robert Lewandowski
2. Luis Suarez
3. Gonzalo Higuain

```python
# Create score
fifa['Gunner'] = (b*fifa['Balance'] + c*fifa['Jumping'] + b*fifa['Ball control'] + d*fifa['Heading accuracy'] + b*fifa['Curve'] + c*fifa['Dribbling'] + d*fifa['Finishing'] + d*fifa['Long shots'] + c*fifa['Positioning'] + a*fifa['Reactions'] + a*fifa['Short passing'] + d*fifa['Shot power'] + a*fifa['Sprint speed'] + c*fifa['Acceleration'] + b*fifa['Vision'] + c*fifa['Strength'] + d*fifa['Volleys']) / (3*a + 4*b + 5*c + 5*d)
# Get the top 3 strikers
fifa.loc[fifa['Gunner'].dropna().sort_values(ascending=False).index][(fifa['Preferred Positions'] == 'ST') | (fifa['Preferred Positions'] == 'RF') | (fifa['Preferred Positions'] == 'LF') | (fifa['Preferred Positions'] == 'CF')]['Name'].head(3)
```
